```python
"""
Advanced Diabetes Risk Prediction with Enhanced Augmentation, Hyperparameter Tuning & XAI
===========================================================================================
Complete ML Pipeline with 1000x Augmentation, Grid Search, Deep Learning, and SHAP
"""

# ============================================================================
# 1Ô∏è‚É£ ENVIRONMENT SETUP & LIBRARY IMPORTS
# ============================================================================

# Install required packages
!pip install imbalanced-learn shap scikit-learn pandas numpy matplotlib seaborn xgboost lightgbm optuna -q

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import (RandomForestClassifier, GradientBoostingClassifier,
                              AdaBoostClassifier, ExtraTreesClassifier)
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, roc_auc_score, confusion_matrix,
                             classification_report, roc_curve, auc)
from imblearn.over_sampling import SMOTE, ADASYN
from imblearn.combine import SMOTEENN
import shap
import warnings
warnings.filterwarnings('ignore')
from datetime import datetime
import time

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (14, 8)
plt.rcParams['font.size'] = 10

print("=" * 80)
print("üöÄ ADVANCED DIABETES PREDICTION SYSTEM WITH XAI")
print("=" * 80)
print("‚úÖ All libraries imported successfully!")
print(f"‚è∞ Started at: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print("=" * 80)

# ============================================================================
# 2Ô∏è‚É£ DATA LOADING & COMPREHENSIVE EXPLORATION
# ============================================================================

filename = "/kaggle/input/300-people/Data of 300 people v2.csv"
df = pd.read_csv(filename)

print(f"\n‚úÖ Dataset loaded: {filename}")
print(f"üìä Shape: {df.shape[0]} rows √ó {df.shape[1]} columns")
print("\n" + "=" * 80)

# Comprehensive Data Analysis
print("üîç COMPREHENSIVE DATA ANALYSIS")
print("=" * 80)

# Basic info
print("\nüìã Dataset Information:")
print(df.info())

# Statistical summary
print("\nüìä Statistical Summary:")
print(df.describe())

# Missing values
print("\n‚ö†Ô∏è Missing Values:")
missing = df.isnull().sum()
if missing.sum() > 0:
    print(missing[missing > 0])
else:
    print("‚úÖ No missing values detected")

# Target distribution
print("\nüéØ Target Variable Distribution:")
print(df['diabetes_status'].value_counts())
print("\nProportions:")
print(df['diabetes_status'].value_counts(normalize=True).round(4))

# ============================================================================
# 3Ô∏è‚É£ ADVANCED DATA VISUALIZATION (FIXED ‚Äî NO OVERLAP)
# ============================================================================

print("\n" + "=" * 80)
print("üìä GENERATING EXPLORATORY DATA ANALYSIS VISUALIZATIONS ‚Äî FIXED SPACING")
print("=" * 80)

import seaborn as sns
sns.set_style("whitegrid")

fig = plt.figure(figsize=(24, 18))
gs = fig.add_gridspec(4, 3, hspace=0.45, wspace=0.35)

# 1. Target Distribution
ax1 = fig.add_subplot(gs[0, 0])
target_counts = df['diabetes_status'].value_counts()
colors = plt.cm.Set3(range(len(target_counts)))
ax1.bar(target_counts.index, target_counts.values, color=colors, edgecolor='black')
ax1.set_title('Target Variable Distribution', fontweight='bold')
ax1.set_xlabel('Diabetes Status')
ax1.set_ylabel('Count')
ax1.tick_params(axis='x', rotation=0)

# 2. Gender Distribution
ax2 = fig.add_subplot(gs[0, 1])
gender_counts = df['gender'].value_counts()
ax2.pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%',
        colors=plt.cm.Pastel1(range(len(gender_counts))), startangle=90, textprops={'fontsize': 10})
ax2.set_title('Gender Distribution', fontweight='bold')

# 3. Age Distribution
ax3 = fig.add_subplot(gs[0, 2])
age_counts = df['age'].value_counts().sort_index()
ax3.bar(age_counts.index.astype(str), age_counts.values,
       color='steelblue', edgecolor='black')
ax3.set_title('Age Group Distribution', fontweight='bold')
ax3.set_xlabel('Age Group')
ax3.set_ylabel('Count')
ax3.tick_params(axis='x', rotation=45)

# 4. Marital Status
ax4 = fig.add_subplot(gs[1, 0])
marital_counts = df['marital_status'].value_counts()
ax4.barh(marital_counts.index, marital_counts.values, color='coral', edgecolor='black')
ax4.set_title('Marital Status Distribution', fontweight='bold')
ax4.set_xlabel('Count')

# 5. Education Level
ax5 = fig.add_subplot(gs[1, 1])
edu_counts = df['education'].value_counts()
ax5.bar(edu_counts.index.astype(str), edu_counts.values,
        color='lightgreen', edgecolor='black')
ax5.set_title('Education Distribution', fontweight='bold')
ax5.tick_params(axis='x', rotation=45)

# 6. Daily Exercise
ax6 = fig.add_subplot(gs[1, 2])
exercise_counts = df['daily_exercise'].value_counts()
ax6.bar(exercise_counts.index.astype(str), exercise_counts.values,
        color='gold', edgecolor='black')
ax6.set_title('Daily Exercise Duration', fontweight='bold')
ax6.tick_params(axis='x', rotation=45)

# 7. Correlation Heatmap
ax7 = fig.add_subplot(gs[2, :])
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
if len(num_cols) > 1:
    corr_matrix = df[num_cols].corr()
    sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', center=0,
                square=True, linewidths=0.5, ax=ax7,
                cbar_kws={"shrink": 0.6})
    ax7.set_title('Feature Correlation Heatmap', fontweight='bold')

# 8. Diabetes Status by Gender
ax8 = fig.add_subplot(gs[3, 0])
sns.countplot(x='gender', hue='diabetes_status', data=df, ax=ax8,
              palette=['#ff9999', '#66b3ff'])
ax8.set_title('Diabetes Status by Gender', fontweight='bold')
ax8.set_xlabel('Gender')
ax8.set_ylabel('Count')
ax8.tick_params(axis='x', rotation=0)

# 9. Diabetes Status by Age Group
ax9 = fig.add_subplot(gs[3, 1])
sns.countplot(x='age', hue='diabetes_status', data=df, ax=ax9,
              palette=['#ff9999', '#66b3ff', '#99ff99'])
ax9.set_title('Diabetes Status by Age Group', fontweight='bold')
ax9.set_xlabel('Age Group')
ax9.set_ylabel('Count')
ax9.tick_params(axis='x', rotation=45)

# 10. Family Diabetes
ax10 = fig.add_subplot(gs[3, 2])
family_counts = df['family_diabetes'].value_counts()
ax10.pie(family_counts.values, labels=family_counts.index,
         autopct='%1.1f%%', colors=['#ff6b6b', '#4ecdc4'],
         startangle=90, textprops={'fontsize': 10})
ax10.set_title('Family Diabetes History', fontweight='bold')

plt.tight_layout(pad=2.0)  # ‚úÖ ensures no overlap
plt.savefig('comprehensive_eda_clean.png', dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Clean EDA visualizations saved as 'comprehensive_eda_clean.png'")


# ============================================================================
# 4Ô∏è‚É£ DATA PREPROCESSING
# ============================================================================

print("\n" + "=" * 80)
print("‚öôÔ∏è ADVANCED DATA PREPROCESSING")
print("=" * 80)

df_processed = df.copy()

# Handle missing values
df_processed = df_processed.fillna(df_processed.mode().iloc[0])

# Separate features and target
target_col = 'diabetes_status'
X = df_processed.drop(columns=[target_col])
y = df_processed[target_col]

# Identify column types
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()

print(f"\nüìä Feature Analysis:")
print(f"   - Total Features: {X.shape[1]}")
print(f"   - Categorical: {len(categorical_cols)}")
print(f"   - Numerical: {len(numerical_cols)}")

# Encode categorical variables
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col].astype(str))
    label_encoders[col] = le

# Encode target
le_target = LabelEncoder()
y_encoded = le_target.fit_transform(y)

print(f"‚úÖ Encoding completed")
print(f"‚úÖ Target classes: {list(le_target.classes_)}")

# ============================================================================
# 5Ô∏è‚É£ TRAIN-TEST-VALIDATION SPLIT
# ============================================================================

print("\n" + "=" * 80)
print("üì¶ STRATEGIC DATA SPLITTING (70-15-15)")
print("=" * 80)

X_train, X_temp, y_train, y_temp = train_test_split(
    X, y_encoded, test_size=0.30, random_state=42, stratify=y_encoded
)

X_test, X_val, y_test, y_val = train_test_split(
    X_temp, y_temp, test_size=0.50, random_state=42, stratify=y_temp
)

print(f"üìä Training:   {X_train.shape[0]} samples ({X_train.shape[0]/len(X)*100:.1f}%)")
print(f"üìä Testing:    {X_test.shape[0]} samples ({X_test.shape[0]/len(X)*100:.1f}%)")
print(f"üìä Validation: {X_val.shape[0]} samples ({X_val.shape[0]/len(X)*100:.1f}%)")

# ============================================================================
# 6Ô∏è‚É£ ADVANCED DATA AUGMENTATION (1000x)
# ============================================================================

print("\n" + "=" * 80)
print("üîÑ MASSIVE DATA AUGMENTATION (1000x TARGET)")
print("=" * 80)

print(f"\nüìä Original training size: {len(X_train)} samples")

# Calculate target size
target_size = len(X_train) * 1000

# Get class distribution
unique, counts = np.unique(y_train, return_counts=True)
n_classes = len(unique)

print(f"üéØ Target size: {target_size} samples")
print(f"üìä Original class distribution:")
for u, c in zip(unique, counts):
    print(f"   Class {le_target.classes_[u]}: {c} samples")

# Multi-stage augmentation strategy
print("\nüîÑ Stage 1: SMOTE augmentation...")
smote = SMOTE(random_state=42, k_neighbors=min(5, min(counts)-1))
X_stage1, y_stage1 = smote.fit_resample(X_train, y_train)
print(f"   ‚úÖ Stage 1 complete: {len(X_stage1)} samples")

# Stage 2: Gaussian noise injection
print("üîÑ Stage 2: Gaussian noise augmentation...")
X_stage1_array = X_stage1.values if isinstance(X_stage1, pd.DataFrame) else X_stage1
augmented_X = [X_stage1_array]
augmented_y = [y_stage1]

# Calculate how many iterations needed
samples_per_round = len(X_stage1)
needed_samples = target_size - samples_per_round
iterations = max(1, needed_samples // samples_per_round)

print(f"   üîÑ Applying {iterations} augmentation iterations...")

for i in range(min(iterations, 50)):  # Cap at 50 iterations for memory
    # Add Gaussian noise (small std to preserve patterns)
    noise = np.random.normal(0, 0.05, X_stage1_array.shape)
    X_noisy = X_stage1_array + noise

    augmented_X.append(X_noisy)
    augmented_y.append(y_stage1)

    if (i + 1) % 10 == 0:
        current_size = len(augmented_X) * samples_per_round
        print(f"      Progress: {current_size:,} samples generated...")

# Combine all augmented data
X_train_augmented = np.vstack(augmented_X)
y_train_augmented = np.concatenate(augmented_y)

# Shuffle
shuffle_idx = np.random.permutation(len(X_train_augmented))
X_train_augmented = X_train_augmented[shuffle_idx]
y_train_augmented = y_train_augmented[shuffle_idx]

print(f"\n‚úÖ Augmentation Complete!")
print(f"   üìä Final training size: {len(X_train_augmented):,} samples")
print(f"   üìà Augmentation factor: {len(X_train_augmented)/len(X_train):.1f}x")

unique_aug, counts_aug = np.unique(y_train_augmented, return_counts=True)
print(f"\nüìä Augmented class distribution:")
for u, c in zip(unique_aug, counts_aug):
    print(f"   Class {le_target.classes_[u]}: {c:,} samples")

# ============================================================================
# 7Ô∏è‚É£ FEATURE SCALING
# ============================================================================

print("\n" + "=" * 80)
print("üìè FEATURE SCALING")
print("=" * 80)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_augmented)
X_test_scaled = scaler.transform(X_test)
X_val_scaled = scaler.transform(X_val)

print("‚úÖ StandardScaler applied to all datasets")

# ============================================================================
# 8Ô∏è‚É£ EXPANDED MODEL TRAINING WITH HYPERPARAMETER TUNING
# ============================================================================

print("\n" + "=" * 80)
print("ü§ñ ADVANCED MODEL TRAINING WITH HYPERPARAMETER TUNING")
print("=" * 80)

# Define models with custom hyperparameters
models = {
    'Logistic Regression': {
        'model': LogisticRegression(random_state=42, max_iter=2000),
        'params': {
            'C': [0.1, 1, 10],
            'penalty': ['l2'],
            'solver': ['lbfgs', 'liblinear']
        }
    },
    'Random Forest': {
        'model': RandomForestClassifier(random_state=42),
        'params': {
            'n_estimators': [100, 200, 300],
            'max_depth': [10, 15, 20],
            'min_samples_split': [2, 5],
            'min_samples_leaf': [1, 2]
        }
    },
    'XGBoost': {
        'model': XGBClassifier(random_state=42, eval_metric='logloss'),
        'params': {
            'n_estimators': [100, 200],
            'max_depth': [5, 7, 10],
            'learning_rate': [0.01, 0.1, 0.3],
            'subsample': [0.8, 1.0]
        }
    },
    'LightGBM': {
        'model': LGBMClassifier(random_state=42, verbose=-1),
        'params': {
            'n_estimators': [100, 200],
            'max_depth': [5, 10, 15],
            'learning_rate': [0.01, 0.1],
            'num_leaves': [31, 50]
        }
    },
    'Gradient Boosting': {
        'model': GradientBoostingClassifier(random_state=42),
        'params': {
            'n_estimators': [100, 200],
            'max_depth': [3, 5, 7],
            'learning_rate': [0.01, 0.1],
            'subsample': [0.8, 1.0]
        }
    },
    'Extra Trees': {
        'model': ExtraTreesClassifier(random_state=42),
        'params': {
            'n_estimators': [100, 200],
            'max_depth': [10, 15, 20],
            'min_samples_split': [2, 5]
        }
    },
    'AdaBoost': {
        'model': AdaBoostClassifier(random_state=42),
        'params': {
            'n_estimators': [50, 100, 200],
            'learning_rate': [0.01, 0.1, 1.0]
        }
    },
    'K-Nearest Neighbors': {
        'model': KNeighborsClassifier(),
        'params': {
            'n_neighbors': [3, 5, 7, 11],
            'weights': ['uniform', 'distance'],
            'metric': ['euclidean', 'manhattan']
        }
    },
    'SVM': {
        'model': SVC(probability=True, random_state=42),
        'params': {
            'C': [0.1, 1, 10],
            'kernel': ['rbf', 'linear'],
            'gamma': ['scale', 'auto']
        }
    },
    'Naive Bayes': {
        'model': GaussianNB(),
        'params': {
            'var_smoothing': [1e-9, 1e-8, 1e-7]
        }
    }
}

# Train models with GridSearchCV
trained_models = {}
results = []
best_params_dict = {}

print(f"\nüîÑ Training {len(models)} models with hyperparameter tuning...")
print("   (This may take several minutes due to large dataset and grid search)\n")

for name, config in models.items():
    print(f"üîÑ Training {name}...")
    start_time = time.time()

    try:
        # Grid Search with cross-validation
        grid_search = GridSearchCV(
            config['model'],
            config['params'],
            cv=3,  # 3-fold CV for speed
            scoring='accuracy',
            n_jobs=-1,
            verbose=0
        )

        grid_search.fit(X_train_scaled, y_train_augmented)
        best_model = grid_search.best_estimator_
        trained_models[name] = best_model
        best_params_dict[name] = grid_search.best_params_

        # Predictions
        y_train_pred = best_model.predict(X_train_scaled)
        y_test_pred = best_model.predict(X_test_scaled)
        y_val_pred = best_model.predict(X_val_scaled)

        # Metrics
        train_acc = accuracy_score(y_train_augmented, y_train_pred)
        test_acc = accuracy_score(y_test, y_test_pred)
        val_acc = accuracy_score(y_val, y_val_pred)

        test_precision = precision_score(y_test, y_test_pred, average='weighted', zero_division=0)
        test_recall = recall_score(y_test, y_test_pred, average='weighted', zero_division=0)
        test_f1 = f1_score(y_test, y_test_pred, average='weighted', zero_division=0)

        # AUC
        if hasattr(best_model, 'predict_proba'):
            y_test_proba = best_model.predict_proba(X_test_scaled)
            if len(np.unique(y_test)) == 2:
                test_auc = roc_auc_score(y_test, y_test_proba[:, 1])
            else:
                test_auc = roc_auc_score(y_test, y_test_proba,
                                        multi_class='ovr', average='weighted')
        else:
            test_auc = None

        elapsed = time.time() - start_time

        results.append({
            'Model': name,
            'Train Acc': train_acc,
            'Test Acc': test_acc,
            'Val Acc': val_acc,
            'Precision': test_precision,
            'Recall': test_recall,
            'F1-Score': test_f1,
            'AUC': test_auc,
            'Time (s)': elapsed
        })

        print(f"   ‚úÖ {name}: Train={train_acc:.4f} | Test={test_acc:.4f} | Val={val_acc:.4f} | Time={elapsed:.1f}s")

    except Exception as e:
        print(f"   ‚ùå {name} failed: {str(e)}")
        continue

# ============================================================================
# 9Ô∏è‚É£ COMPREHENSIVE PERFORMANCE ANALYSIS
# ============================================================================

print("\n" + "=" * 80)
print("üìä MODEL PERFORMANCE COMPARISON")
print("=" * 80)

results_df = pd.DataFrame(results)
results_df = results_df.sort_values('Val Acc', ascending=False)
print("\n" + results_df.to_string(index=False))

# Save results
results_df.to_csv('detailed_model_performance.csv', index=False)

# Find best model
best_model_name = results_df.iloc[0]['Model']
best_model = trained_models[best_model_name]

print(f"\n{'='*80}")
print(f"üèÜ BEST MODEL: {best_model_name}")
print(f"{'='*80}")
print(f"   üìä Validation Accuracy: {results_df.iloc[0]['Val Acc']:.4f}")
print(f"   üìä Test Accuracy: {results_df.iloc[0]['Test Acc']:.4f}")
print(f"   üìä F1-Score: {results_df.iloc[0]['F1-Score']:.4f}")
print(f"   ‚öôÔ∏è Best Parameters: {best_params_dict[best_model_name]}")

# ============================================================================
# üîü ADVANCED VISUALIZATIONS
# ============================================================================

print("\n" + "=" * 80)
print("üìä GENERATING ADVANCED RESULT VISUALIZATIONS")
print("=" * 80)

# Create comprehensive visualization
fig = plt.figure(figsize=(20, 16))
gs = fig.add_gridspec(4, 3, hspace=0.35, wspace=0.3)

# 1. Model Accuracy Comparison
ax1 = fig.add_subplot(gs[0, :])
x = np.arange(len(results_df))
width = 0.25
ax1.bar(x - width, results_df['Train Acc'], width, label='Train', alpha=0.8)
ax1.bar(x, results_df['Test Acc'], width, label='Test', alpha=0.8)
ax1.bar(x + width, results_df['Val Acc'], width, label='Validation', alpha=0.8)
ax1.set_xlabel('Models', fontweight='bold', fontsize=12)
ax1.set_ylabel('Accuracy', fontweight='bold', fontsize=12)
ax1.set_title('Model Accuracy Comparison (Train/Test/Validation)', fontweight='bold', fontsize=14)
ax1.set_xticks(x)
ax1.set_xticklabels(results_df['Model'], rotation=45, ha='right')
ax1.legend(fontsize=10)
ax1.grid(axis='y', alpha=0.3)
ax1.axhline(y=0.8, color='r', linestyle='--', alpha=0.5, label='80% Threshold')

# 2. Performance Metrics Heatmap
ax2 = fig.add_subplot(gs[1, :])
metrics_data = results_df[['Model', 'Test Acc', 'Precision', 'Recall', 'F1-Score']].set_index('Model')
sns.heatmap(metrics_data.T, annot=True, fmt='.3f', cmap='RdYlGn',
            center=0.75, vmin=0.5, vmax=1.0, ax=ax2, cbar_kws={'label': 'Score'})
ax2.set_title('Performance Metrics Heatmap (Test Set)', fontweight='bold', fontsize=14)
ax2.set_xlabel('')
ax2.set_ylabel('Metrics', fontweight='bold', fontsize=12)

# 3. Confusion Matrix
ax3 = fig.add_subplot(gs[2, 0])
y_best_pred = best_model.predict(X_test_scaled)
cm = confusion_matrix(y_test, y_best_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax3,
            xticklabels=le_target.classes_, yticklabels=le_target.classes_,
            cbar_kws={'label': 'Count'})
ax3.set_xlabel('Predicted', fontweight='bold', fontsize=11)
ax3.set_ylabel('Actual', fontweight='bold', fontsize=11)
ax3.set_title(f'Confusion Matrix - {best_model_name}', fontweight='bold', fontsize=12)

# 4. ROC Curve
ax4 = fig.add_subplot(gs[2, 1])
if hasattr(best_model, 'predict_proba'):
    y_test_proba = best_model.predict_proba(X_test_scaled)

    if len(np.unique(y_test)) == 2:
        fpr, tpr, _ = roc_curve(y_test, y_test_proba[:, 1])
        roc_auc = auc(fpr, tpr)
        ax4.plot(fpr, tpr, linewidth=2, label=f'{best_model_name} (AUC = {roc_auc:.3f})')
    else:
        for i in range(len(le_target.classes_)):
            fpr, tpr, _ = roc_curve((y_test == i).astype(int), y_test_proba[:, i])
            roc_auc = auc(fpr, tpr)
            ax4.plot(fpr, tpr, linewidth=2, label=f'{le_target.classes_[i]} (AUC={roc_auc:.3f})')

    ax4.plot([0, 1], [0, 1], 'k--', linewidth=1, label='Random')
    ax4.set_xlabel('False Positive Rate', fontweight='bold', fontsize=11)
    ax4.set_ylabel('True Positive Rate', fontweight='bold', fontsize=11)
    ax4.set_title(f'ROC Curve - {best_model_name}', fontweight='bold', fontsize=12)
    ax4.legend(fontsize=9)
    ax4.grid(alpha=0.3)

# 5. Training Time Comparison
ax5 = fig.add_subplot(gs[2, 2])
ax5.barh(results_df['Model'], results_df['Time (s)'], color='skyblue', edgecolor='black')
ax5.set_xlabel('Training Time (seconds)', fontweight='bold', fontsize=11)
ax5.set_title('Model Training Time', fontweight='bold', fontsize=12)
ax5.grid(axis='x', alpha=0.3)

# 6. Precision vs Recall
ax6 = fig.add_subplot(gs[3, 0])
for idx, row in results_df.iterrows():
    ax6.scatter(row['Recall'], row['Precision'], s=200, alpha=0.6)
    ax6.annotate(row['Model'], (row['Recall'], row['Precision']),
                fontsize=8, ha='center', va='center')
ax6.set_xlabel('Recall', fontweight='bold', fontsize=11)
ax6.set_ylabel('Precision', fontweight='bold', fontsize=11)
ax6.set_title('Precision vs Recall Trade-off', fontweight='bold', fontsize=12)
ax6.grid(alpha=0.3)

# 7. F1-Score Ranking
ax7 = fig.add_subplot(gs[3, 1])
sorted_f1 = results_df.sort_values('F1-Score', ascending=True)
colors = plt.cm.RdYlGn(np.linspace(0.3, 0.9, len(sorted_f1)))
ax7.barh(sorted_f1['Model'], sorted_f1['F1-Score'], color=colors, edgecolor='black')
ax7.set_xlabel('F1-Score', fontweight='bold', fontsize=11)
ax7.set_title('F1-Score Ranking', fontweight='bold', fontsize=12)
ax7.axvline(x=0.75, color='r', linestyle='--', alpha=0.5)
ax7.grid(axis='x', alpha=0.3)

# 8. Model Performance Radar
ax8 = fig.add_subplot(gs[3, 2], projection='polar')
categories = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
best_values = [
    results_df.iloc[0]['Test Acc'],
    results_df.iloc[0]['Precision'],
    results_df.iloc[0]['Recall'],
    results_df.iloc[0]['F1-Score']
]
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
best_values += best_values[:1]
angles += angles[:1]
ax8.plot(angles, best_values, 'o-', linewidth=2, label=best_model_name)
ax8.fill(angles, best_values, alpha=0.25)
ax8.set_xticks(angles[:-1])
ax8.set_xticklabels(categories, fontsize=10)
ax8.set_ylim(0, 1)
ax8.set_title(f'Performance Radar - {best_model_name}', fontweight='bold',
              fontsize=12, pad=20)
ax8.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))
ax8.grid(True)

plt.savefig('comprehensive_model_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Comprehensive visualizations saved as 'comprehensive_model_analysis.png'")

# ============================================================================
# 1Ô∏è‚É£1Ô∏è‚É£ EXPLAINABLE AI - SHAP ANALYSIS (FIXED)
# ============================================================================

print("\n" + "=" * 80)
print("üîç EXPLAINABLE AI - SHAP ANALYSIS")
print("=" * 80)

print(f"\nüîÑ Generating SHAP explanations for {best_model_name}...")

# Sample data for SHAP (for computational efficiency with large datasets)
sample_size = min(500, len(X_test_scaled))
X_test_sample = X_test_scaled[:sample_size]
y_test_sample = y_test[:sample_size]

print(f"   Using {sample_size} test samples for SHAP analysis...")

# Create SHAP explainer based on model type
if best_model_name in ['Random Forest', 'Extra Trees', 'XGBoost', 'LightGBM',
                        'Gradient Boosting', 'AdaBoost']:
    explainer = shap.TreeExplainer(best_model)
    shap_values = explainer.shap_values(X_test_sample)
else:
    # Use KernelExplainer for other models
    background = shap.sample(X_train_scaled, 100)
    explainer = shap.KernelExplainer(best_model.predict_proba, background)
    shap_values = explainer.shap_values(X_test_sample)

print("‚úÖ SHAP values calculated successfully")

# Handle multi-class SHAP values
if isinstance(shap_values, list):
    # Multi-class: use class 1 or aggregate
    if len(shap_values) == 2:
        shap_values_plot = shap_values[1]
    else:
        # For multi-class, use the first class
        shap_values_plot = shap_values[0]
else:
    shap_values_plot = shap_values

# SHAP Visualizations
fig, axes = plt.subplots(3, 1, figsize=(16, 18))

# 1. SHAP Summary Plot (Beeswarm)
plt.sca(axes[0])
shap.summary_plot(shap_values_plot, X_test_sample,
                  feature_names=X.columns.tolist(),
                  show=False, max_display=20)
axes[0].set_title('SHAP Summary Plot - Feature Impact Distribution',
                  fontweight='bold', fontsize=14, pad=20)

# 2. SHAP Bar Plot (Mean Importance)
plt.sca(axes[1])
shap.summary_plot(shap_values_plot, X_test_sample,
                  feature_names=X.columns.tolist(),
                  plot_type="bar", show=False, max_display=20)
axes[1].set_title('SHAP Feature Importance - Top 20 Features',
                  fontweight='bold', fontsize=14, pad=20)

# 3. SHAP Waterfall Plot (Single Prediction)
plt.sca(axes[2])
shap.plots._waterfall.waterfall_legacy(explainer.expected_value[1] if isinstance(explainer.expected_value, np.ndarray) else explainer.expected_value,
                                        shap_values_plot[0],
                                        feature_names=X.columns.tolist(),
                                        max_display=15,
                                        show=False)
axes[2].set_title('SHAP Waterfall Plot - Single Prediction Explanation',
                  fontweight='bold', fontsize=14, pad=20)

plt.tight_layout()
plt.savefig('shap_analysis_comprehensive.png', dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ SHAP visualizations saved as 'shap_analysis_comprehensive.png'")

# Calculate feature importance (FIXED)
print("\nüìä Calculating feature importance...")

# Fix for the error: ensure 1D array
if len(shap_values_plot.shape) > 1:
    mean_shap = np.abs(shap_values_plot).mean(axis=0)
else:
    mean_shap = np.abs(shap_values_plot)

# Create feature importance dataframe
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': mean_shap
}).sort_values('Importance', ascending=False)

print("\n" + "=" * 80)
print("üìä TOP 15 MOST IMPORTANT FEATURES (SHAP)")
print("=" * 80)
print(feature_importance.head(15).to_string(index=False))

# Save feature importance
feature_importance.to_csv('feature_importance_shap.csv', index=False)

# Additional SHAP visualization - Feature importance with values
fig, ax = plt.subplots(figsize=(14, 10))
top_features = feature_importance.head(20)
colors = plt.cm.viridis(np.linspace(0.3, 0.9, len(top_features)))
bars = ax.barh(top_features['Feature'], top_features['Importance'],
               color=colors, edgecolor='black', linewidth=1.5)
ax.set_xlabel('Mean |SHAP Value|', fontweight='bold', fontsize=12)
ax.set_title('Top 20 Feature Importance Rankings (SHAP)',
             fontweight='bold', fontsize=14, pad=20)
ax.invert_yaxis()
ax.grid(axis='x', alpha=0.3)

# Add value labels
for i, (bar, val) in enumerate(zip(bars, top_features['Importance'])):
    ax.text(val + 0.01, bar.get_y() + bar.get_height()/2,
            f'{val:.3f}', va='center', fontweight='bold', fontsize=9)

plt.tight_layout()
plt.savefig('feature_importance_ranking.png', dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Feature importance ranking saved as 'feature_importance_ranking.png'")

# ============================================================================
# 1Ô∏è‚É£2Ô∏è‚É£ DETAILED CLASSIFICATION REPORT
# ============================================================================

print("\n" + "=" * 80)
print("üìã DETAILED CLASSIFICATION REPORT")
print("=" * 80)

y_test_pred = best_model.predict(X_test_scaled)
y_val_pred = best_model.predict(X_val_scaled)

print(f"\n{'='*80}")
print(f"TEST SET PERFORMANCE - {best_model_name}")
print(f"{'='*80}")
print(classification_report(y_test, y_test_pred,
                          target_names=[str(c) for c in le_target.classes_],
                          digits=4))

print(f"\n{'='*80}")
print(f"VALIDATION SET PERFORMANCE - {best_model_name}")
print(f"{'='*80}")
print(classification_report(y_val, y_val_pred,
                          target_names=[str(c) for c in le_target.classes_],
                          digits=4))

# ============================================================================
# 1Ô∏è‚É£3Ô∏è‚É£ CREATIVE ADDITION: PREDICTION CONFIDENCE ANALYSIS
# ============================================================================

print("\n" + "=" * 80)
print("üéØ CREATIVE ANALYSIS: PREDICTION CONFIDENCE DISTRIBUTION")
print("=" * 80)

if hasattr(best_model, 'predict_proba'):
    y_test_proba = best_model.predict_proba(X_test_scaled)
    max_probas = np.max(y_test_proba, axis=1)

    # Analyze confidence
    high_confidence = np.sum(max_probas > 0.8)
    medium_confidence = np.sum((max_probas > 0.6) & (max_probas <= 0.8))
    low_confidence = np.sum(max_probas <= 0.6)

    print(f"\nüìä Prediction Confidence Breakdown (Test Set):")
    print(f"   üü¢ High Confidence (>80%): {high_confidence} samples ({high_confidence/len(max_probas)*100:.1f}%)")
    print(f"   üü° Medium Confidence (60-80%): {medium_confidence} samples ({medium_confidence/len(max_probas)*100:.1f}%)")
    print(f"   üî¥ Low Confidence (<60%): {low_confidence} samples ({low_confidence/len(max_probas)*100:.1f}%)")

    # Visualization
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # Confidence histogram
    axes[0].hist(max_probas, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
    axes[0].axvline(x=0.8, color='green', linestyle='--', linewidth=2, label='High Confidence')
    axes[0].axvline(x=0.6, color='orange', linestyle='--', linewidth=2, label='Medium Confidence')
    axes[0].set_xlabel('Prediction Confidence', fontweight='bold', fontsize=12)
    axes[0].set_ylabel('Frequency', fontweight='bold', fontsize=12)
    axes[0].set_title('Prediction Confidence Distribution', fontweight='bold', fontsize=14)
    axes[0].legend(fontsize=11)
    axes[0].grid(alpha=0.3)

    # Confidence by prediction correctness
    correct_pred = (y_test == y_test_pred)
    correct_conf = max_probas[correct_pred]
    incorrect_conf = max_probas[~correct_pred]

    axes[1].hist([correct_conf, incorrect_conf], bins=20,
                 color=['green', 'red'], alpha=0.6,
                 label=['Correct Predictions', 'Incorrect Predictions'],
                 edgecolor='black')
    axes[1].set_xlabel('Prediction Confidence', fontweight='bold', fontsize=12)
    axes[1].set_ylabel('Frequency', fontweight='bold', fontsize=12)
    axes[1].set_title('Confidence: Correct vs Incorrect Predictions',
                     fontweight='bold', fontsize=14)
    axes[1].legend(fontsize=11)
    axes[1].grid(alpha=0.3)

    plt.tight_layout()
    plt.savefig('prediction_confidence_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()

    print("‚úÖ Confidence analysis saved as 'prediction_confidence_analysis.png'")

# ============================================================================
# 1Ô∏è‚É£4Ô∏è‚É£ CREATIVE ADDITION: LEARNING CURVE ANALYSIS
# ============================================================================

print("\n" + "=" * 80)
print("üìà CREATIVE ANALYSIS: LEARNING CURVE")
print("=" * 80)

from sklearn.model_selection import learning_curve

print("üîÑ Computing learning curve (this may take a moment)...")

# Use a subset for learning curve computation
train_sizes = np.linspace(0.1, 1.0, 10)
train_sizes_abs, train_scores, val_scores = learning_curve(
    best_model, X_train_scaled[:5000], y_train_augmented[:5000],
    train_sizes=train_sizes, cv=3, scoring='accuracy',
    n_jobs=-1, random_state=42
)

train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
val_mean = np.mean(val_scores, axis=1)
val_std = np.std(val_scores, axis=1)

fig, ax = plt.subplots(figsize=(12, 7))
ax.plot(train_sizes_abs, train_mean, 'o-', color='blue', linewidth=2,
        label='Training Score')
ax.fill_between(train_sizes_abs, train_mean - train_std,
                train_mean + train_std, alpha=0.2, color='blue')
ax.plot(train_sizes_abs, val_mean, 'o-', color='red', linewidth=2,
        label='Cross-Validation Score')
ax.fill_between(train_sizes_abs, val_mean - val_std,
                val_mean + val_std, alpha=0.2, color='red')
ax.set_xlabel('Training Set Size', fontweight='bold', fontsize=12)
ax.set_ylabel('Accuracy Score', fontweight='bold', fontsize=12)
ax.set_title(f'Learning Curve - {best_model_name}', fontweight='bold', fontsize=14)
ax.legend(loc='lower right', fontsize=11)
ax.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('learning_curve.png', dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Learning curve saved as 'learning_curve.png'")

# ============================================================================
# 1Ô∏è‚É£5Ô∏è‚É£ MODEL SAVING & EXPORT
# ============================================================================

print("\n" + "=" * 80)
print("üíæ SAVING RESULTS AND MODEL ARTIFACTS")
print("=" * 80)

# Save all results
results_df.to_csv('complete_model_results.csv', index=False)
feature_importance.to_csv('feature_importance_shap.csv', index=False)

# Save best parameters
import json
with open('best_hyperparameters.json', 'w') as f:
    json.dump(best_params_dict, f, indent=4)

print("‚úÖ Files saved:")
print("   üìÑ complete_model_results.csv")
print("   üìÑ feature_importance_shap.csv")
print("   üìÑ best_hyperparameters.json")
print("   üñºÔ∏è comprehensive_eda.png")
print("   üñºÔ∏è comprehensive_model_analysis.png")
print("   üñºÔ∏è shap_analysis_comprehensive.png")
print("   üñºÔ∏è feature_importance_ranking.png")
print("   üñºÔ∏è prediction_confidence_analysis.png")
print("   üñºÔ∏è learning_curve.png")

# ============================================================================
# 1Ô∏è‚É£6Ô∏è‚É£ EXECUTIVE SUMMARY
# ============================================================================

print("\n" + "=" * 80)
print("üéØ EXECUTIVE SUMMARY")
print("=" * 80)

import json, time
from datetime import datetime

# Safely access best result
best_result = results_df.iloc[0]

# ‚úÖ Safe formatting for AUC (handles None or NaN)
auc_value = (
    f"{best_result['AUC']:.4f}"
    if best_result.get('AUC') not in [None, float('nan')] and not isinstance(best_result['AUC'], str)
    else "N/A"
)

# ‚úÖ Safe high confidence
high_confidence_value = (
    high_confidence if 'high_confidence' in locals() and hasattr(best_model, 'predict_proba') else "N/A"
)

# ‚úÖ Safe hyperparameters
best_hyperparams = (
    json.dumps(best_params_dict.get(best_model_name, {}), indent=4)
    if isinstance(best_params_dict, dict)
    else "N/A"
)

# ‚úÖ Safe feature importance
top_features_text = ""
if "feature_importance" in locals() and not feature_importance.empty:
    for i, row in feature_importance.head(5).iterrows():
        top_features_text += f"‚îú‚îÄ {i+1}. {row['Feature']}: {row['Importance']:.4f}\n"
else:
    top_features_text = "‚îú‚îÄ Feature importance data unavailable.\n"

# ‚úÖ Compute generalization & performance interpretation
gen_quality = (
    "Good"
    if abs(best_result['Train Acc'] - best_result['Val Acc']) < 0.1
    else "Needs Improvement"
)
performance_quality = (
    "Excellent" if best_result['Val Acc'] > 0.85
    else "Good" if best_result['Val Acc'] > 0.75
    else "Moderate"
)

# ‚úÖ Main Summary
summary = f"""
‚ïî‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïó
‚ïë                        DIABETES PREDICTION PROJECT                            ‚ïë
‚ïë                          EXECUTIVE SUMMARY REPORT                             ‚ïë
‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù

üìä DATASET OVERVIEW
‚îú‚îÄ Original Samples: {df.shape[0]:,}
‚îú‚îÄ Features: {df.shape[1] - 1}
‚îú‚îÄ Target Classes: {len(le_target.classes_)}
‚îî‚îÄ Augmented Training Samples: {len(X_train_augmented):,} ({len(X_train_augmented)/len(X_train):.0f}x augmentation)

ü§ñ MODEL TRAINING
‚îú‚îÄ Total Models Trained: {len(models)}
‚îú‚îÄ Hyperparameter Tuning: GridSearchCV with 3-fold CV
‚îú‚îÄ Training Strategy: 70% Train / 15% Test / 15% Validation
‚îî‚îÄ Augmentation Method: SMOTE + Gaussian Noise Injection

üèÜ BEST MODEL: {best_model_name}
‚îú‚îÄ Training Accuracy: {best_result['Train Acc']:.4f} ({best_result['Train Acc']*100:.2f}%)
‚îú‚îÄ Test Accuracy: {best_result['Test Acc']:.4f} ({best_result['Test Acc']*100:.2f}%)
‚îú‚îÄ Validation Accuracy: {best_result['Val Acc']:.4f} ({best_result['Val Acc']*100:.2f}%)
‚îú‚îÄ Precision: {best_result['Precision']:.4f}
‚îú‚îÄ Recall: {best_result['Recall']:.4f}
‚îú‚îÄ F1-Score: {best_result['F1-Score']:.4f}
‚îú‚îÄ AUC Score: {auc_value}
‚îî‚îÄ Training Time: {best_result['Time (s)']:.2f} seconds

‚öôÔ∏è BEST HYPERPARAMETERS
{best_hyperparams}

üîç TOP 5 MOST IMPORTANT FEATURES (SHAP Analysis)
{top_features_text}
üìà MODEL INSIGHTS
‚îú‚îÄ High Confidence Predictions (>80%): {high_confidence_value}
‚îú‚îÄ Model Generalization: {gen_quality}
‚îî‚îÄ Overall Performance: {performance_quality}

‚úÖ DELIVERABLES
‚îú‚îÄ Trained Models: {len(trained_models)} models
‚îú‚îÄ Visualizations: 6 comprehensive plots
‚îú‚îÄ SHAP Analysis: Complete feature importance & interpretability
‚îî‚îÄ Performance Reports: Detailed metrics for all models

üéâ PROJECT STATUS: COMPLETED SUCCESSFULLY
‚è∞ Total Execution Time: {(time.time() - start_time) / 60:.2f} minutes

‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê
"""

# ‚úÖ Print and save summary
print(summary)

with open('executive_summary.txt', 'w') as f:
    f.write(summary)

print("‚úÖ Executive summary saved as 'executive_summary.txt'")
print("\n" + "=" * 80)
print("üéä ALL TASKS COMPLETED SUCCESSFULLY!")
print("=" * 80)
print("\nüì¶ Ready for deployment and stakeholder presentation!")
print("üöÄ Project execution finished at:", datetime.now().strftime('%Y-%m-%d %H:%M:%S'))
print("=" * 80)
```
